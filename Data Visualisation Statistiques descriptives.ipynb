# Visualisation d’un nuage de points et exploration des variables

## import , ouverture et lecture de fichier csv

```python
import pandas as pd
import seaborn as sb
df = pd.read_csv('Downloads/house_pricing.csv')

print(df)
                     
```

### identification des types de données

```python
df.dtypes
```

### nuage de points (lotArea x SalePrice)

```python
df.plot.scatter(x = 'LotArea', y = 'SalePrice');
```

### filtrage en fonction de lotArea et SalePrice 

```python
bc = df[(df['LotArea'] < 20000) & (df['SalePrice'] < 500000) ]
sb.scatterplot(data=bc, y="LotArea", x="SalePrice");
```

### Nuage de point (LotArea x LotFrontage) 

```python
sb.scatterplot(data=df, y="LotArea", x="LotFrontage");
```

### filtrage en fonction de LotArea et LotFrontage

```python
fil2 = df[(df['LotArea'] < 100000) & (df['LotFrontage'] < 200) ]
sb.scatterplot(data=fil2,  x="LotFrontage",y="LotArea");
```

```python
#df.plot.scatter(x = 'LotFrontage', y = 'LotArea');
from matplotlib import pyplot as plt
sb.scatterplot(data=df, x="LotFrontage", y="LotArea", hue="SalePrice");
```
